```python
def main():
    numbers = []
    total_sum = 0

    while True:
        try:
          number = float(input("Введите число (или 0 для завершения): "))
        except ValueError:
          print ("введите корректное число")
          continue

        if number == 0:
            break

        numbers.append(number)
        total_sum += number
        if total_sum == 0:
           print("Сумма всех введённых чисел равна 0. ")
           break

    sum_of_squares = sum(x ** 2 for x in numbers)
    print("Сумма квадратов всех введённых чисел:", sum_of_squares)

if __name__ == "__main__":
    main()
```
